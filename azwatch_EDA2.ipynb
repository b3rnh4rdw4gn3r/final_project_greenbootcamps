```python
#Getting necessary libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset
file_path = '/Users/bernhardwagner/Documents/Final Project/data/AZWatch_subscribers.csv'
data = pd.read_csv(file_path)

# Define the custom color palette
custom_palette = ['#C4F7FC', '#3B9491', '#002C26']
custom_palette_adjusted = ['#C4F7FC', '#3B9491']

# Display the first few rows of the dataset
print("\nStatistical summary of the dataset:")
print(data.describe())

# Check for missing values
print("\nNumber of missing values in each column:")
print(data.isnull().sum())
print("---------------------------------")

# Check for duplicate rows
duplicate_rows = data.duplicated()
print("\nNumber of duplicate rows in the dataset:", duplicate_rows.sum())


```

```python

# Boxplots for the numeric columns with custom color palette
plt.figure(figsize=(12, 6))

# Boxplot of Engagement Time
plt.subplot(1, 2, 1)
plt.boxplot(data['engagement_time'], patch_artist=True, boxprops=dict(facecolor=custom_palette[0]))
plt.title('Boxplot of Engagement Time', color=custom_palette[0])
plt.ylabel('Engagement Time')
# Assumption: Shows the distribution and variability of user engagement time, highlighting outliers and typical engagement duration.

# Boxplot of Engagement Frequency
plt.subplot(1, 2, 2)
plt.boxplot(data['engagement_frequency'], patch_artist=True, boxprops=dict(facecolor=custom_palette[1]))
plt.title('Boxplot of Engagement Frequency', color=custom_palette[1])
plt.ylabel('Engagement Frequency')
# Assumption: Illustrates the spread and central tendency of engagement frequency, indicating common engagement patterns.

plt.tight_layout()
plt.show()

```

### Meaning
1) Shows the distribution and variability of user engagement time, highlighting outliers and typical engagement duration.
2) Illustrates the spread and central tendency of engagement frequency, indicating common engagement patterns.


```python
# Distplots for the numeric columns with custom color palette
plt.figure(figsize=(12, 6))

# Distribution of Engagement Time
plt.subplot(1, 2, 1)
sns.histplot(data['engagement_time'], kde=True, color=custom_palette[0])
plt.title('Distribution of Engagement Time', color=custom_palette[0])
plt.xlabel('Engagement Time')
plt.ylabel('Frequency')
# Assumption: Highlights the common engagement times and overall distribution, useful for identifying typical user engagement duration.

# Distribution of Engagement Frequency
plt.subplot(1, 2, 2)
sns.histplot(data['engagement_frequency'], kde=True, color=custom_palette[1])
plt.title('Distribution of Engagement Frequency', color=custom_palette[1])
plt.xlabel('Engagement Frequency')
plt.ylabel('Frequency')
# Assumption: Shows how often users engage with the platform, helping to identify frequent vs. infrequent users.

plt.tight_layout()
plt.show()

# Bar plot for age group and subscription status
plt.figure(figsize=(10, 6))
sns.countplot(data=data, x='age_group', hue='subscription_status', palette=custom_palette_adjusted)
plt.title('Subscription Status by Age Group', color=custom_palette[1])
plt.xlabel('Age Group')
plt.ylabel('Count')
plt.legend(title='Subscription Status')
# Assumption: Compares subscription statuses across age groups, revealing which age groups have higher or lower subscription rates.

plt.tight_layout()
plt.show()
```

### Meaning
1) Highlights the common engagement times and overall distribution, useful for identifying typical user engagement duration.
2) Shows how often users engage with the platform, helping to identify frequent vs. infrequent users.
3) Compares subscription statuses across age groups, revealing which age groups have higher or lower subscription rates.

```python

# Scatter plot with histograms for examining relationships between numerical variables and subscription status
fig, axes = plt.subplots(2, 2, figsize=(14, 12))

# Distribution of Engagement Time by Subscription Status
sns.histplot(data=data, x='engagement_time', hue='subscription_status', kde=True, palette=custom_palette_adjusted, ax=axes[0, 0])
axes[0, 0].set_title('Distribution of Engagement Time by Subscription Status')
axes[0, 0].set_xlabel('Engagement Time')
axes[0, 0].set_ylabel('Frequency')

# Distribution of Engagement Frequency by Subscription Status
sns.histplot(data=data, x='engagement_frequency', hue='subscription_status', kde=True, palette=custom_palette_adjusted, ax=axes[0, 1])
axes[0, 1].set_title('Distribution of Engagement Frequency by Subscription Status')
axes[0, 1].set_xlabel('Engagement Frequency')
axes[0, 1].set_ylabel('Frequency')

# Scatter Plot: Engagement Time vs. Engagement Frequency
sns.scatterplot(data=data, x='engagement_time', y='engagement_frequency', hue='subscription_status', palette=custom_palette_adjusted, ax=axes[1, 0])
axes[1, 0].set_title('Engagement Time vs. Engagement Frequency')
axes[1, 0].set_xlabel('Engagement Time')
axes[1, 0].set_ylabel('Engagement Frequency')
axes[1, 0].legend(title='Subscription Status')

# Removing the unused subplot
fig.delaxes(axes[1, 1])

plt.tight_layout()
plt.show()
```

### Meaning 
Peaks in the distribution indicate common engagement times among subscribed and churned users, useful for retention strategies.
Assumption: Identifies common engagement frequencies, helping to segment users and predict churn.
Reveals the relationship between engagement time and frequency, indicating how these metrics correlate with subscription status.
